```python
import initialize
import mistify
import torch.nn as nn
import torch
from torch.utils import data as torch_data
```

```python
class Net(nn.Module):

    def __init__(self):
        super().__init__()
        self.and1 = mistify.infer.MinSum(8, 16)
        self.or1 = mistify.infer.MaxProd(16, 4)
        self.and2 = mistify.infer.MinSum(4, 1)

    def forward(self, x: torch.Tensor) -> torch.Tensor:

        y = self.and1(x)
        y = self.or1(y)
        return self.and2(y)
```

```python

X = torch.rand(2000, 8)
net_b = Net()
T = net_b(X).detach()
net = Net()

optim = torch.optim.Adam(net.parameters(), lr=1e-3)
dataset = torch_data.TensorDataset(X, T)


for i in range(1000):
    for x_i, t_i in torch_data.DataLoader(dataset, 128, True):

        optim.zero_grad()

        y_i = net(x_i)
        loss = (y_i - t_i).pow(2).mean()
        print(loss.item())
        loss.backward()
        optim.step()
```
